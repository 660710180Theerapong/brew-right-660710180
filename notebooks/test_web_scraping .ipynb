```python
from bs4 import BeautifulSoup
```

```python
def fetch_coffee_prices():
    headers = {
        'User-Agent': 'Mozilla/5.0 (compatible; Brew-Right Bot 1.0;)'
    }
    url = 'https://pantip.com/tag/Starbucks'
    
    try:
        resp = requests.get(url, headers=headers, timeout=10)
        resp.raise_for_status()
        soup = BeautifulSoup(resp.content, 'html.parser')

        prices = []
        for row in soup.find_all('li'):
            if 'กาแฟ' in row.get_text():
                cols = row.find_all('a')
                if len(cols) >= 3:
                    prices.append({
                        'product': cols[0].get_text(strip=True),
                        'unit': cols[1].get_text(strip=True),
                        'price': cols[2].get_text(strip=True),
                        'scraped_at': datetime.now().isoformat(),
                        'source_url': url
                    })
        
        return prices
    except Exception as e:
        print(f"Error: {e}")
        return []

# ดึงข้อมูล
coffee_prices = fetch_coffee_prices()
coffee_prices
```
